# Centrality 

### Degree Centrality

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()
# G.add_edges_from([('c', 'a'), ('a', 'b'), ('b', 'd'), ('d', 'e'), ('e', 'g'), ('e', 'f')])
with open("graph1.txt", "r") as file:
    for line in file:
        edge = line.strip().split()
        if len(edge) == 2:
            G.add_edge(edge[0], edge[1])

# Compute degree centrality
degree_centrality = nx.degree_centrality(G)
print("Degree Centrality:", degree_centrality)

# Rank nodes by degree centrality
ranked_nodes = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)

rankings = {}
rank = 1
prev_centrality = None

for node, centrality in ranked_nodes:
    if prev_centrality is not None and centrality != prev_centrality:
        rank += 1
    
    rankings[node] = rank  #
    prev_centrality = centrality

# Print the rankings
print("Nodes ranked by degree centrality:")
print(rankings)

# Draw the graph
nx.draw(G, with_labels=True, node_color='lightblue', node_size=700, font_size=10)
plt.show()
```

### K-Shell Centrality

```python
import networkx as nx
import matplotlib.pyplot as plt

def min_degree(G):
    return min(G.degree, key=lambda x: x[1])[1]

def create_bucket(G, degree):
    return [node for node in G.nodes if G.degree(node) == degree]

def remove_nodes(G, degree):
    nodes_to_remove = [node for node in G.nodes if G.degree(node) == degree]
    G.remove_nodes_from(nodes_to_remove)
    return G

def k_shell_centrality(G):
    dictionary_ans = {}
    prev_min_deg = -1
    while len(G.nodes) > 0:
        min_deg = min_degree(G)
        print(min_deg)
        bucket = create_bucket(G, min_deg)
        print(bucket)
        prev_min_deg = max(min_deg, prev_min_deg)
        
        if prev_min_deg not in dictionary_ans:
            dictionary_ans[prev_min_deg] = []
        
        dictionary_ans[prev_min_deg].extend(bucket)
        G = remove_nodes(G, min_deg)
    return dictionary_ans

G = nx.Graph()
G.add_edges_from(
    [(1, 2), (1, 9), (3, 13), (4, 6),
     (5, 6), (5, 7), (5, 8), (5, 9), 
     (5, 10), (5, 11), (5, 12), (10, 12), 
     (10, 13), (11, 14), (12, 14), 
     (12, 15), (13, 14), (13, 15), 
     (13, 17), (14, 15), (15, 16)]
)

k_shell_result = k_shell_centrality(G.copy())

print("K-Shell Centrality Result:")
# for k, nodes in k_shell_result.items():
#     print(f"Shell {k}: Nodes {nodes}")
print(k_shell_result)

nx.draw(G, with_labels=True, node_color='red', node_size=700, font_size=10)
plt.show()
```

### STC Centrality

```python
import networkx as nx
from collections import deque
import matplotlib.pyplot as plt

def bfs_upto_level(graph, start_node, max_level):
    visited = set()
    result = []
    queue = deque([(start_node, 0)])
    
    while queue:
        node, level = queue.popleft()
        
        if level > max_level:
            break
        
        if node not in visited:
            visited.add(node)
            result.append((node, level))
            
            for neighbor in graph.neighbors(node):
                if neighbor not in visited:
                    queue.append((neighbor, level + 1))
    
    return result

def total_triangles_in_bfs(graph, start_node, max_level):
    nodes_in_bfs = bfs_upto_level(graph, start_node, max_level)
    # print(start_node+' has nodes : ',nodes_in_bfs)
    triangles = nx.triangles(graph)  
    total = 0
    
    for node, level in nodes_in_bfs:
        total += triangles[node] / pow(2, level+1) 
    
    return total

def stcCentrality(graph):
    centrality = {}
    for node in graph.nodes():
        centrality[node] = total_triangles_in_bfs(graph, node, 3)  
    return centrality

G = nx.Graph()

# G.add_edges_from([(0, 1), (1, 2), (2, 0), (3, 1), (2,3)])
with open("graph2.txt", "r") as file:
    for line in file:
        edge = line.strip().split()
        if len(edge) == 2:
            G.add_edge(edge[0], edge[1])


ans=stcCentrality(G);
print('STC Centrality : ',ans)

ranked_nodes = sorted(ans.items(), key=lambda x: x[1], reverse=True)

rankings = {}
rank = 1
prev_centrality = None

for node, centrality in ranked_nodes:
    if prev_centrality is not None and centrality != prev_centrality:
        rank += 1
    
    rankings[node] = rank  
    prev_centrality = centrality

# Print the rankings
print("Nodes ranked by STC Centrality:")
print(rankings)


nx.draw(G, with_labels=True, node_color='lightblue', node_size=500, font_size=15)
plt.show()
```

### K-Sum Centrality

```python
import networkx as nx
from collections import deque
import matplotlib.pyplot as plt

def bfs_upto_level(graph, start_node, max_level):
    visited = set()
    result = []
    queue = deque([(start_node, 0)])
    
    while queue:
        node, level = queue.popleft()
        
        if level > max_level:
            break
        
        if node not in visited:
            visited.add(node)
            result.append((node, level))
            
            for neighbor in graph.neighbors(node):
                if neighbor not in visited:
                    queue.append((neighbor, level + 1))
    
    return result

def total_degree_in_bfs(graph, start_node, max_level):
    nodes_in_bfs = bfs_upto_level(graph, start_node, max_level)
    total = 0
    
    for node, level in nodes_in_bfs:
        total += graph.degree(node)
    
    return total

def ksumCentrality(graph):
    centrality = {}
    for node in graph.nodes():
        centrality[node] = total_degree_in_bfs(graph, node, 2)  
    return centrality

G = nx.Graph()

# G.add_edges_from([(0, 1), (1, 2), (2, 0), (3, 1), (2,3)])
with open("graph2.txt", "r") as file:
    for line in file:
        edge = line.strip().split()
        if len(edge) == 2:
            G.add_edge(edge[0], edge[1])



ans=ksumCentrality(G)
print('KSum Centrality : ',ans)

ranked_nodes = sorted(ans.items(), key=lambda x: x[1], reverse=True)

rankings = {}
rank = 1
prev_centrality = None

for node, centrality in ranked_nodes:
    if prev_centrality is not None and centrality != prev_centrality:
        rank += 1
    
    rankings[node] = rank  
    prev_centrality = centrality

# Print the rankings
print("Nodes ranked by KSum Centrality:")
print(rankings)

nx.draw(G, with_labels=True, node_color='red', node_size=500, font_size=15)
plt.show()
```
